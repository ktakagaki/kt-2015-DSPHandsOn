### Extraction of triangle with highest amplitude and template matching

```python
def spikeDTW(template,data):
    """ 
    Calculates the warping costs for a template and input sequence.
    Specilly for Spikes. Ensures that the depolarization and repolarization are compared and the costs are
    separate calculated and then added.
    
    Parameters
    ----------
    template: array_like
    
    data: array_like
    
    Returns
    -------
    totalCost: double
        Total cost for warping both arrays.
    """
    
    if type(template) != list:
        maxValueTemplate = template.tolist().index(max(template))
    else:
        maxValueTemplate = template.index(max(template))
        
    if type(data) != list:
        maxValueData = data.tolist().index(max(data))
    else: 
        maxValueData = data.index(max(data))
        
        
    costmatrix = np.zeros((len(template),len(data)))
    cumulativeCostmatrix = np.zeros((len(template),len(data)))
    
    
    #Calculation of the Costmatrix
    for i in range(0,maxValueTemplate):
        for j in range(0,maxValueData):
            costmatrix[i,j] = abs(template[i]-data[j])
            
    for i in range(maxValueTemplate,len(template)):
        for j in range(maxValueData,len(data)):
            costmatrix[i,j] = abs(template[i]-data[j])
            
    #Calculation of Cumulative Costmatrix
    for i in range (maxValueTemplate,-1,-1):
        for j in range(maxValueData,-1,-1):
            
            if i == maxValueTemplate and j == maxValueData:
                cumulativeCostmatrix[i,j] = costmatrix[i,j]
                
            elif i == maxValueTemplate and j != maxValueData:
                cumulativeCostmatrix[i,j] = costmatrix[i,j] + cumulativeCostmatrix[i,j+1]
            
            elif i != maxValueTemplate and j == maxValueData:
                cumulativeCostmatrix[i,j] = costmatrix[i,j] + cumulativeCostmatrix[i+1,j]
                
            elif i != maxValueTemplate and j != maxValueData:
                cumulativeCostmatrix[i,j] = costmatrix[i,j] + min(cumulativeCostmatrix[i+1,j],cumulativeCostmatrix[i,j+1],cumulativeCostmatrix[i+1,j+1])
                                                                  
                
    for i in range (maxValueTemplate,len(template)):
        for j in range(maxValueData,len(data)):
            
            if i == maxValueTemplate and j == maxValueData:
                cumulativeCostmatrix[i,j] = costmatrix[i,j]
                
            elif i == maxValueTemplate and j != maxValueData:
                cumulativeCostmatrix[i,j] = costmatrix[i,j] + cumulativeCostmatrix[i,j-1]
            
            elif i != maxValueTemplate and j == maxValueData:
                cumulativeCostmatrix[i,j] = costmatrix[i,j] + cumulativeCostmatrix[i-1,j]
                
            elif i != maxValueTemplate and j != maxValueData:
                cumulativeCostmatrix[i,j] = costmatrix[i,j] + min(cumulativeCostmatrix[i-1,j],cumulativeCostmatrix[i,j-1],cumulativeCostmatrix[i-1,j-1])           
            
    totalCost = cumulativeCostmatrix[0,0] + cumulativeCostmatrix[len(template)-1, len(data)-1]      
    
    return totalCost
```

```python
def shape(data, threshold):
    """ Extract different Spike shapes compare all and classify them.
        
        Parameters
        ----------
        data: dictionary
            Contains each recorded signal as an array with a unique key.
    
        thresholds: dictionary
            Contains all calculated thresholds belonging to a signal.
    
        Returns
        -------
        shapes: dictionary
            Contains extracted shapes and their class.
            
        positionlist: array
            All postions where spike amplitudes where detected. 
    """
    
    count  = -1
    count2 = 0
    timer = 0
    positionlist = []
    templates_checked = []
    total_cost = []
    shapes = {}
    
    #Extract the recorded signals one by one.
    for v in new_data:
        temp = new_data[v]
        
        #Check all elements in the Signal if they passed the threshold. 
        for element in temp:
            count += 1 
            if element >= threshold["threshold{0}".format(v[-1])] and (count - 10) >= 0 and (count + 14) <= len(temp) and timer == 0:
                tempdata = [w for w in temp[count -10 : count + 14]]
                tempsavedcounter = np.linspace(count-10, count + 14, 25)
                maxposition = tempdata.index(max(tempdata))
                position = tempsavedcounter[maxposition]
                tempdata = [w for w in temp[position -10 : position + 14]]
                #Control all other signals at the same position if one of them contains a spike with a higher amplitude. 
                for i in new_data:
                    if i[-1] != v[-1]:
                        tempdata2 = new_data[i]
                        tempdata2 = tempdata2[position -10 : position + 14]
                        if max(tempdata2) > max(tempdata):
                            tempdata = tempdata2
                #Store the position where a spike was localised. 
                position = int(position)
                
                #If the dictionary is empty, mark the first spike(or noise) as template 1. 
                if bool(not shapes):
                    timer = 16
                    count2 += 1
                    matched_template_counter = 1
                    shapes["spike{0}".format(count2)] = (tempdata,matched_template_counter)
                    positionlist.append(position)
                #All the following detected spikes are compared with each template once. Similarity is always calculated 
                #with the spike wich was first assigned to a class(template-group).    
                elif all([i not in positionlist for i in range(position -6, position +10)]) and bool(shapes):
                    count2 += 1
                    for key in shapes:
                        if shapes[key][1] not in templates_checked:
                            totalcost = spikeDTW(shapes[key][0],tempdata)
                            total_cost.append(totalcost)
                            templates_checked.append(shapes[key][1])
                    #When the costs between all templates and the new detected spike bigger then 5 are,
                    #mark the current spike as a new template(5 is just for this easy example and must be adjusted for real signals)
                    #If there is one template with costs below 5, add the spike to this class.
                    if min(total_cost) >= 5:
                        matched_template_counter += 1
                        shapes["spike{0}".format(count2)] = (tempdata,matched_template_counter)
                    else:    
                        matched_template = total_cost.index(min(total_cost))
                        shapes["spike{0}".format(count2)] = (tempdata,matched_template+1)
                            
                    positionlist.append(position)
                    total_cost =[]
                    templates_checked=[]
                    
                    
            elif timer > 0: 
                timer -= 1
        count = -1
    return shapes, positionlist
```

```python
shapes, positionlist = shape(new_data, new_threshold)
```

```python
match1 = 0
match2 = 0
match3 = 0
match4 = 0

for key in shapes:
    if shapes[key][1] == 1:
        match1 += 1
    elif shapes[key][1] == 2:
        match2 += 1
    elif shapes[key][1] == 3:
        match3 += 1
    elif shapes[key][1] == 4:
        match4 += 1
print "Template 1 found: " + str(match1) + " times"
print "Template 2 found: " + str(match2) + " times"
print "Template 3 found: " + str(match3) + " time"
print "Template 4 found: " + str(match4) + " time"
```
